```python
from torch.utils.data import Dataset, DataLoader
import torch
import numpy as np
import random
import os

from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

```python
datasets_dir = "./Data/CombinedData/"
panda = True

if panda:
    print(f"Panda ...")
    filename = os.path.join(datasets_dir+"combined_data_testing_panda.pickle")
    model_path= "./Models/panda_ft_classifier"
else:
    print(f"ILoSA ...")
    filename = os.path.join(datasets_dir+"combined_data_testing_ilosa.pickle")    
    print(f"Ilosa testing dataset loaded")
    model_path = "./Models/ilosa_ft_classifier"
```

```python
data = np.array(np.load(filename, allow_pickle=True,))
```

```python
model = torch.load(model_path)
model.eval()
```

```python
i = random.randint(0, data.shape[0]-1)

X, y = data[i][1], data[i][0]

if y == "right":
    y = 1
else:
    y = 0
    
X = np.array(X, dtype=np.float32).T[None, ]
print(X.shape)
```

```python
X = torch.tensor(X)
y = torch.tensor(y)

_, prediction = torch.max(model(X), 1)

print(f"Prediction: {prediction}")
print(f"Expected output: {y}")
```
